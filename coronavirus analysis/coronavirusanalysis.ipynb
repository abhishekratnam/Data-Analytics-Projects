```python
import pandas as pd
data = pd.read_csv("../input/corona/covid_19_data.csv")
```

```python
confirmed_mean = data.iloc[:,5].mean()
confirmed_std =  data.iloc[:,5].std()
deaths_mean = data.iloc[:,6].mean()
deaths_std = data.iloc[:,6].std()
recovered_mean = data.iloc[:,7].mean()
recovered_std = data.iloc[:,7].std()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set()
plt.title("Deaths ")
data['Deaths'].hist(bins = 20)
```

```python
from collections import Counter
each_country_cases = Counter(data['Province/State'].values)
```

```python
each_country_cases
```

```python
count = 0
for i in each_country_cases.values():
    if i > count:
        count += i
    elif i== count:
        break
print("Max cases found in a country",count)
        
```
